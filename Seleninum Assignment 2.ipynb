# 3 Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
# obtaining the web page details with the given url
driver = webdriver.Chrome("D:\Inernship\chromedriver.exe")
url='https://images.google.com/'
urls = []    # empty list for image urls to download
data = []    # empty list for storing data
search_item = ["Fruits", "Cars", "Machine Learning", "Guitar", "Cakes"]
for item in search_item:
    driver.get(url)  # performing a Google search for our selected items
    time.sleep(3)
    search_bar = driver.find_element_by_tag_name("input")     # finding search bar on Google page   
    search_bar.send_keys(str(item))      # send search item list    
    search_button =driver.find_element_by_xpath("//button[@class='Tg7LZd']").click() # now we will click on search button
    
    # scroll the page in order to generate more images on the website
    for _ in range(100):
        driver.execute_script("window.scrollBy(0,10)")        
        imgs = driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")
    img_url = []
    for image in imgs:
        source = image.get_attribute('src')
        if source is not None:
                if(source[0:4] == 'http'):
                    img_url.append(source)
    for i in img_url[:10]:      # getting only 10 images each
        urls.append(i)
                    
for i in range(len(urls)):
    if i >= 50:
        break
    print("Obtaining {0} of {1} images".format(i, 50))
    response=requests.get(urls[i])
    file = open(r"D:\Inernship\Q3\image"+str(i)+".jpg", "wb")
    file.write(response.content)
```

```python
#6
#importing webdriver
driver=webdriver.Chrome("D:\Inernship\chromedriver.exe")
time.sleep(4)
url="https://trak.in/india-startup-funding-investment-2015/"
driver.get(url)  #getting the funding page
time.sleep(4)
```

```python
#creating empty boxex
date=[]
s_name=[]
industry=[]
sub=[]
location=[]
i_name=[]
i_type=[]
amount=[]
```

```python
#getting the data of the month from july to september
for i in range(54,57):
    # Date
    dt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[2]'.format(i))
    for d in dt:
        date.append(d.text)
        
    #startupname
    s=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[3]'.format(i))
    for d in s:
        s_name.append(d.text)

    #industry
    a=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[4]'.format(i))
    for d in a:
        industry.append(d.text)
    #sub
    su=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[5]'.format(i))
    for d in su:
        sub.append(d.text)
    #location
    lo=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[6]'.format(i))
    for d in lo:
        location.append(d.text)
    #i_name
    ina=driver.find_elements_by_xpath("//*[@id='tablepress-{}']/tbody/tr/td[7]".format(i))
    for d in ina:
        i_name.append(d.text)
    #i_type
    ty=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[8]'.format(i))
    for d in ty:
        i_type.append(d.text)
    #amount
    am=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[9]'.format(i))
    for d in am:
        amount.append(d.text)

time.sleep(2)
```

```python
#creating dataframe
ans6=pd.DataFrame({})
ans6["Date"]=date
ans6["Startup_Name"]=s_name
ans6["Industry"]=industry
ans6["Sub"]=sub
ans6["Location"]=location
ans6["Investors name"]=i_name
ans6["Investment_type"]=i_type
ans6["Amount"]=amount
ans6
```

You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
driver = webdriver.Chrome('D:\Inernship\chromedriver.exe')
# obtaining the web page from the given url
url = "https://www.hostelworld.com/"
driver.get(url)
```

```python
# now we will find the search bar and search for hostels in London location
search_bar = driver.find_element_by_xpath("//input[@id='location-text-input-field']")
search_bar.send_keys('London')
```

```python
button = driver.find_element_by_xpath("//input[@class='search-input']")
button.click()
```

```python
search_button = driver.find_element_by_xpath("//button[@id='search-button']")
search_button.click()
```

```python
# creating empty lists to collect the required information
hostel_name = []
distance = []
pvt_prices = []
dorms_price = []
rating = []
reviews = []
over_all = []
facilities = []
description =[]
product_url = []

for i in driver.find_elements_by_xpath("//div[@class = 'pagination-item pagination-current' or @class='pagination-item']"):
    i.click()
    time.sleep(4)
    
    # fetching hostel name
    try:
        name = driver.find_elements_by_xpath("//h2[@class='title title-6']")
        for i in name:
            hostel_name.append(i.text)
    except NoSuchElementException:
        hostel_name.append('-')
        
    # fetching the distance from city centre    
    try:
        dist = driver.find_elements_by_xpath("//div[@class='subtitle body-3']//a//span[1]")
        for i in dist:
            distance.append(i.text.replace('Hostel - ',''))
    except NoSuchElementException:
        distance.append('-')
        
    for i in driver.find_elements_by_xpath("//div[@class='prices-col']"):
        
    # fetching the privates from price details
        try:
            pvt_price = driver.find_element_by_xpath("//a[@class='prices']//div[1]//div")
            pvt_prices.append(pvt_price.text)
        except NoSuchElementException:
            pvt_prices.append('-')
            
    # fetching the dorms from price information
    for i in driver.find_elements_by_xpath("//div[@class='prices-col']"):
        try:
            dorms = driver.find_element_by_xpath("//a[@class='prices']//div[2]//div")
            dorms_price.append(dorms.text)
        except NoSuchElementException:
            dorms_price.append('-')
            
    # fetching all the facilities
    try:
        fac1 = driver.find_elements_by_xpath("//div[@class='has-wifi']")
        fac2 = driver.find_elements_by_xpath("//div[@class='has-sanitation']")
        for i in fac1:
            for j in fac2:
                facilities.append(i.text +', '+ j.text )
    except NoSuchElementException:
        facilities.append('-')
        
    # fetching the urls of each hostel
    p_url = driver.find_elements_by_xpath("//div[@class='prices-col']//a[2]")
    for i in p_url:
        product_url.append(i.get_attribute('href'))

for i in product_url:
    driver.get(i)
    time.sleep(3)
    
    # clicking on the show more button for description
    try:
        driver.find_element_by_xpath("//a[@class='toggle-content']").click()
        time.sleep(5)
    except NoSuchElementException:
        pass
    
    # fetching the ratings for the hostels
    try:
        rat = driver.find_element_by_xpath("//div[@class='score orange big' or @class='score gray big']")
        rating.append(rat.text)
    except NoSuchElementException:
        rating.append('-')
        
    # fetching the total reviews        
    try:
        rws = driver.find_element_by_xpath("//div[@class='reviews']")
        reviews.append(rws.text.replace('Total Reviews',''))
    except NoSuchElementException:
        reviews.append('-')
        
    # fetching the overall review
    try:
        overall_rw = driver.find_element_by_xpath("//div[@class='keyword']//span")
        over_all.append(overall_rw.text)
    except NoSuchElementException:
        over_all.append('-')
        
    # fetching the property description 
    try:
        disc = driver.find_element_by_xpath("//div[@class='content']")
        description.append(disc.text)
    except NoSuchElementException:
        over_all.append('-')

# now creating a pandas dataframe to show all the collected data
data = list(zip(hostel_name,distance,rating,reviews,over_all,pvt_prices,dorms_price,facilities,description))       
df = pd.DataFrame(data, columns = ["Hostel name","Distance from city centre","ratings","Total reviews",
                                   "Overall review","Privates from price","Dorms from price","Facilities","Property Description"])
df
```
